```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
np.set_printoptions(threshold=np.inf)

files = ["code.xlsx", "knowledge.xlsx", "whole.xlsx"]

for file in files:
    path = "data/" + file
    df = pd.read_excel(path)

    models = ["ChatGPT 4o", "Llama3", "Codestral", "ReservoirChatBasic", "ReservoirChatLittle", "ReservoirChatMedium", "ReservoirChatBig", "NotebookLMCore", "NotebookLMOther", "NotebookLMBig"]
    model_names = ["ChatGPT 4o", "Llama3", "Codestral", "ReservoirChat Basic", "ReservoirChat Little", "ReservoirChat Medium", "ReservoirChat Big", "NotebookLM Core", "NotebookLM Other", "NotebookLM Big"]
    array = []

    for model in models:
        array += [df[model].apply(lambda x: eval(x)).values.sum()]

    array = np.array(array)
    corr = np.corrcoef(array)

    plt.figure(figsize=(8, 6), dpi=80)
    sns.heatmap(corr, annot=True, cmap="Greys", xticklabels=model_names, yticklabels=model_names)
    plt.savefig(f"{path}{file.split('.')[0]}.pdf", bbox_inches='tight')
```
