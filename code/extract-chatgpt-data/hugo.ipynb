```python
import pandas as pd

df = pd.read_json('conversations.json')
filter_hugo = df.title.apply(lambda x: 'Hugo' in x)
df = df[filter_hugo]
df.title = df.title.apply(lambda x: x.split()[0])
df
```

```python
import os

conversations = []
for index, conv in df.iterrows():
    keys = list(conv.mapping.keys())
    instance = f"## Conversation de {conv.title} avec Doctor Hugo\n"
    for key in keys[3:]:
        role = conv.mapping[key]['message']['author']['role']
        role = 'Doctor Hugo' if role == 'assistant' else conv.title
        instance += f"### {role}  \n"
        instance += f"{conv.mapping[key]['message']['content']['parts'][0]}  \n"
        instance += "\n"
    conversations.append(instance)

directory = 'convs'
if not os.path.exists(directory):
    os.makedirs(directory)

for i, conv in enumerate(conversations):
    with open(f'convs/conv_{i}.md', 'a') as f:
        f.write(conv)
```
